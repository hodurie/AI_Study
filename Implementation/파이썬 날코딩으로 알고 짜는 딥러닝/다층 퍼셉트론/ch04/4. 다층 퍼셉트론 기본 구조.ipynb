# 다층 퍼셉트론 기본 구조

## 다층 퍼셉트론 신경망 구조
입력 층과 출력 층 사이에 1개 이상의 층(은닉 층)이 존재하는 신경망

### 은닉 계층의 수와 폭
은닉 계층은 출력에 직접 드러나지 않는 층으로 설계자가 자유롭게 정할 수 있다.  
은닉 계층의 수를 정하고 그 안에 은닉 계층의 폭을 정할 수 있는데,  
은닉 계층의 폭은 사용한 퍼셉트론의 수로 노드의 수라고도 한다.  



### 비선형 활성화 함수

#### ReLU 함수
ReLU 함수는 음수 입력을 걸러내 0으로 만드는 기능을 제공한다.  

```
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(x, 0)

x_data = np.arange(-4, 4, 0.001)
y_data = relu(x_data)

plt.plot(x_data, y_data)
plt.vlines(0, -.1, 4, colors='skyblue')
```
