# 8. 다섯 가지 정규화 확장


## 부적합과 과적합
딥러닝 모델을 학습하다보면 부적합과 과적합이 발생해 학습을 제대로 하지 못하게 된다.  
이때 **부적합**은 모델이 데이터를 충분히 학습하지 못해 학습용 예제를 제대로 풀지 못하는 것이다.  
**과적합**은 모델이 모든 데이터 수용할 정도로 커 학습용 예제를 모조리 외워 버리는 현상이다.  

부적합인 경우는 모델의 용량을 키우고 학습 횟수를 늘리면 해결이 가능하지만 과적합의 경우는 조금 다르다.  

과적합의 경우 학습 과정에 적당한 제약을 둬 지엽적인 특성을 모두 외워버리지 않게 만들어 해결한다.  
이러한 과정의 예들은 다음과 같다.
1. L2 손실
2. L1 손실
3. 드롭 아웃
4. 잡음 주입
5. 배치 정규화

### L2 손실
**L2 손실**은 절대값이 큰 파라미터에 규제를 주는 기법이다.  
즉, 큰 가중치가 폭주해 작은 가중치의 값들을 무시하지 않게 만들어준다.

$$ P_{L2} = \frac{1}{2}\lambda \sum_{i=1}^{n}w_i^2$$
$$L' = L + \frac{1}{2}\lambda \sum_{i=1}^{n}w_i^2 $$

이때,   
$W$ : $ W = w_1 .\ldots ,w_n $  
$P_{L2}$ : L2 패널티  
$L'$ : L2 손실 기법이 적용된 새로운 손실  
$\lambda$ : L2 손실률

### L1 손실
**L1 손실**도 L2 손실과 마찬가지로 손실 함수에 페널티를 더해 정규화하는 기법이다.  

$$ P_{L1} = \alpha \sum_{i=1}^{n}\vert w_i \vert$$
$$ L' = L + \alpha \sum_{i=1}^{n}\vert w_i \vert$$

이때,  
$\alpha$ : L1 손실률  

L1 손실과 L2 손실의 차이점은 큰 변수의 영향을 줄이느냐 작은 변수의 영향을 줄이느냐 이다.  

L1 손실의 경우 모델에 영향을 끼치지 못하는 가중치를 0 또는 0에 가깝게 만든다.  
L2 손실의 경우 모델에 영향을 많이 끼치는 가중치의 값을 줄여 각 가중치들의 값을  비슷하게 만든다.

### 드롭아웃
드롭아웃은 입력 또는 일부 계층의 노드를 사용 하지 않고 신경망을 학습시키는 규제화 기법이다.   

### 잡음 주입
잡음 주입은 두 계층 사이에 잡음을 삽입하여 출력에 적당한 잡음을 추가해 다음 계층에 전달하는 기법이다.  

잡음 주입의 경우 훈련시에만 사용하고 테스트에 대해선 사용하지 않는다.

### 배치 정규화
배치 정규화는 미니 배치 내 벡터 성분별로 정규화를 수행하는 방식이다.  
이때 정규화는 동일한 선형 변환을 통해 평균 0 표준 편차 1로 만들어주는 것을 말한다.
