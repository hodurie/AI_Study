# 7. 간단한 합성곱 모델

## 다층 퍼셉트론의 문제점과 새로운 구조의 필요성

다층 퍼셉트론 신경망은 입력과 출력 사이에 은닉 계층을 갖고 각 층은 모두 연결 돼 있다.  

모든 층이 연결 돼 있어 입력 벡터 크기가 커지면 그만큼 메모리 부담이 커지고 파라미터의 수가 증가한다. 이는 학습이 느려지게 만드는 요인이 된다.  

또한, 물체가 있는 이미지를 학습할 때 동일한 사물이 다른 위치에 있게 되면 이를 학습하지 못한다. 즉, 위치에 대한 불변성을 나타낼 수 없다. 이러한 문제의식에서 출발한 것이 합성곱 신경망이다.

### 합성곱 계층

합성곱 계층은 **합성곱** 이란 수학 연산에서 힌트를 얻어 고안된 처리 방식이다.  

이미지에 커널이라는 작은 사각형을 움직여가며 각 위치에 대응하는 픽셀들의 곱들을 더해 나타낸다. 앞서 구한 결과에 각 커널에 해당하는 편향값을 합산해 결과값을 출력한다.  

이를 통해 다음과 같은 효과를 낼 수 있다.
- 커널의 반복 이용으로 학습 횟수가 늘어나 빠른 학습이 가능
- 같은 커널을 모든 위치에 적용해 한 곳에서 포착된 패턴이 다른 곳에 이용 할 수 있음



### 합성곱 연산의 패딩과 건너뛰기

합성곱 연산의 패딩은 두 가지가 존재한다.
- SAME
- VALID

#### SAME
SAME의 경우 합성곱을 진행 하고 나서도 입력 이미지의 크기와 출력 이미지의 크기를 동일하게 만드는 방법이다.  

이를 가능하게 만들기 위해 원래 이미지에 상하좌우에 0의 값을 채워넣어 합성곱을 진행한다.

#### VALID
VALID의 경우 이미지 상하좌우에 0값을 채우지 않고 합성곱을 진행하는 방식을 말한다.

#### 합성곱 연산의 건너뛰기
합성곱 연산의 건너뛰기는 커널이 이미지를 얼만큼의 보폭으로 움직일 것인가에 대한 수치로, stride라고도 한다.

### 풀링 계층
풀링은 일정 영역 안에 있는 입력 픽셀값들을 최대치, 평균치 같은 대푯값을 구해 출력 픽셀로 생성하는 처리를 말한다.  
주어진 입력에 대푯값만 출력하는 과정을 거치기 때문에 별도의 파라미터가 필요없다. 덕분에 순전파 및 역전파 처리의 속도가 빠르다.

### 채널의 도입과 커널의 확장

합성곱의 커널은 입력 데이터의 특정 패턴이나 특징에 민감하게 반응해 이를 학습하는 경향이 있다. 하나의 커널만으로는 이미지가 담고있는 여러 패턴들을 모두 담을 수 없기에 커널의 수를 늘려 특징을 잘 담을 수 있도록 잘 조절해줘야 한다.

### 합성곱과 풀링의 역전파 처리

합성곱 계층의 경우 완전 연결층의 역전파 처리와 비슷하며 차이점이라면 4차원의 값을 다뤄야 한다는 것이다.

최대치 풀링의 경우 최댓값으로 선정 된 위치에만 손실 기울기를 전달하고 나머지는 0을 전달한다. 만약 최대치가 여러개 나왔다면 그 중 하나만 골라 손실 기울기를 전해주는 방법을 취한다.

평균치 풀링의 경우 손실 기울기에 원소의 수만큼 나눠 전달하는 방식을 취한다.

### 합성곱 신경망의 일반적인 구성

합성곱 신경망은 보통 합성곱 계층과 풀링 계층을 번갈아 두며 마지막엔 완전 연결층을 연결해 출력층을 만드는 형태로 구성된다.
