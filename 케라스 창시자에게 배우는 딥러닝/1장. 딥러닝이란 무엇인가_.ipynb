
# 1장 딥러닝이란 무엇인가?

## 1.인공지능과 머신러닝, 딥러닝
### 1.1 인공지능

**보통의 사람이 수행하는 지능적인 작업을 자동화하기 위한 연구 활동**  

#### 초기 인공지능  
- 많은 규칙을 제시해 해답을 찾게 하는 방법
- **심볼릭 AI(symbolic AI)**  
- **전문가 시스템**  

#### 한계점  
- 특정 규칙을 찾아내는 것이 힘듦

#### 개선  
- **머신러닝**  

### 1.2 머신 러닝

**경험을 통해 자동으로 개선하는 컴퓨터 알고리즘의 연구**
- 데이터에서 통계적 구조를 찾아 그 작업을 자동화  

#### 머신러닝의 종류
- 나이브 베이즈. 로지스틱 회귀
- 얕은 신경망
- 서포트 벡터 머신
- 결정트리, 랜덤포레스트, 그래디언트 부스팅

#### 머신러닝 구성 요소
- **입력 데이터** 
- **기대 출력**
- **알고리즘 성능 측정 방법**

#### 통계와의 차이점
- 머신러닝은 대량의 복잡한 데이터 셋을 다룸

### 1.3 딥러닝

**데이터 표현을 학습하기 위한 다단계 처리 방식**

#### deep
- 여러 층을 활용하여 표현을 학습
- **층 기반 표현 학습**
- **계층적 표현 학습**

#### 작동 원리
1. 데이터 입력
2. 각 층의 가중치와 결합하여 예측값 도출
3. 손실 함수를 통해 실제값과 예측값과의 차이 계산
4. 옵티마이저를 활용하여 가중치 수정
5. 최소한의 손실을 낼 때까지 위의 과정을 반복

#### 머신러닝과의 차이점
- 여러 layer를 이용
- 특성 공학을 자동화 함
 > 특성 공학이란, 초기 학습을 위한 데이터 변환  
 > 학습이란, 더 나은 표현을 찾는 자동화된 과정  
 > 표현이란, 의미 있는 데이터로의 변환
