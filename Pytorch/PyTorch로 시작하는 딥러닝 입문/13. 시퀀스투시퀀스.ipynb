# PyTorch로 시작하는 딥러닝 입문
## 13. 시퀀스투시퀀스

### 시퀀스투시퀀스
시퀀스-투-시퀀스(Sequence-to-Sequence)는 입력된 시퀀스로부터 다른 도메인의 시퀀스를 출력하는 다양한 분야에서 사용되는 모델로  
챗봇, 내용 요약, STT 등에 사용 됨  

#### 모델의 개요
**seq2seq**는 번역기에서 대표적으로 사용하는 모델  

다음과 같은 과정을 거침  
1. 문자 데이터를 순차적으로 받아 인코더를 통해 하나의 벡터로 만듦  
 - 이때 인코더를 통해 생성된 벡터는 컨텍스트 벡터(context vector)라고 함  
2. 컨텍스트 벡터를 디코더로 전송
3. 디코더는 컨텍스트 벡터를 받아서 번역된 단어를 한 개씩 순차적으로 출력


#### seq2seq의 동작 과정
**seq2seq**의 경우 두 개의 RNN 층(LSTM or GRU)으로 이루어져 있음  
- 인코딩
- 디코딩

인코더 RNN 셀의 마지막 시점의 은닉 상태를 디코더 RNN 셀로 넘겨주는데 이를 컨텍스트 벡터하고,   
컨텍스트 벡터는 디코더 RNN 셀의 첫번째 은닉 상태로 사용

**seq2seq 동작 과정**  
1. 테스트 단계
2. 훈련 단계와 교사 강요
3. 임베딩 층(Embedding layer)
4. RNN 셀
5. 디코더
