# GAN 개념


## Encoder
Encoder의 경우 기존 데이터를 낮은 차원으로 매핑시켜 유용한 정보(잠재 변수)를 추출한다.



##  Decoder
Decoder의 경우 잠재 변수를 기반으로 원래의 차원으로 복구 하는 과정을 말한다.

## 적대적 휸련
실제 이미지셋과 가짜 이미지를 생성해주는 생성기가 존재한다고 가정해보자.  
여기서 진짜와 비슷한 이미지를 생성하고 싶을 때 쓰는 것이 적대적 훈련(GAN) 이다.  

GAN은 다음과 같은 단계로 실행 된다.
1. 판별기에 실제 데이터를 보여주고 1 로 판단하게 한다.
    - 실제 데이터를 보여주고 이미지를 판단하게 하며 이때 발생 한 오차로 판별기를 업데이트한다.  
2. 생성기로 만들어진 가짜 데이터를 0 으로 판단하게 한다.
    - 생성기로 나온 데이터를 보여주고 이미지를 판단하게 하며 발생 한 오차로 판별기만 업데이트 한다.  
3. 생성기의 결과를 보여주고 1 로 판단하게 한다.
    - 생성기로 이미지를 만든 후 분류기가 1 로 판별하게 끔 생성기만을 업데이트한다.
