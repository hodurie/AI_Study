# OpenCV 시작하기

## 이미지 크기
- 이미지의 크기는 너비와 높이를 의미 하며 행렬 형태로 구성 됨
- OpenCV에서 이미지는 (행, 열, 1 or 3), (높이, 너비, 1 or 3)으로 표현

## 정밀도
- 정밀도(Bit Depth)는 이미지가 얼마나 많은 색상을 표현할 수 있는 가를 의미


### 비트 표현
- 비트는 색상의 표현 개수
- n-Bit로 표현 ($2^n$)


## 채널
- 채널은 RGB의 형태로 구성 됨
    - 흑백 이미지의 경우 1개의 채널
    - OpenCV의 채널은 BGR로 구성 됨
- 흑백은 단일 채널을 사용 해야함
    - 다중 채널로 변경 시 동일한 값이 두 번 더 들어감
    - 연산량 증가와 정확도 감소 초래


## 관심 영역
- 관심 영역은 ROI(Region of Interest)로 표현하며 이미지 상에서 관심 있는 영역을 의미


## 관심 채널
- 관심 채널은 COI(Channel of Interest)로 표현하며 이미지 상에 관심 있는 채널을 의미
- 관심 영역과 관심 채널은 특정 영역만 불러와 연산량 감소와 정확도 향상에 쓰임

## 히스토그램
- 도수 분포표로 데이터 분포를 몇 개의 구간으로 나눠 각 구간에 속하는 데이터를 시각적으로 표현한 막대 그래프
- 빈도 수, 차원 수, 범위의 세가지 중요 요소를 갖고 있음
    - 빈도 수 : 히스토그램 그래프 $X$축 간격
    - 차원 수 : 히스토그램을 분석할 이미지 차원
    - 범위 : 히스토그램 그래프의 $X$축 범위
